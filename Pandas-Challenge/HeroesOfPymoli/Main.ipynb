```python
# Dependencies and Setup
import pandas as pd
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#count total
Total_Players=purchase_data["Purchase ID"].count()
Total_Players
#create df
Total_df=pd.DataFrame({"Total Players":[Total_Players]})
Total_df
```

```python
#number of unique items
Unique_items=purchase_data["Item Name"].value_counts()
count_unique=len(Unique_items)
#average price
avg_price=purchase_data["Price"].mean()
#total revenue
total_price=purchase_data["Price"].sum()
#create df
Summary_df=pd.DataFrame({"Number of Unique Items":[count_unique],"Average Purchase Price":[avg_price],
                         "Total Number of Purchases":[Total_Players],"Total Revenue":[total_price]})
#formate df
Summary_df["Average Purchase Price"] = (Summary_df["Average Purchase Price"]).apply(lambda x: '${:,.2f}'.format(x))
Summary_df["Total Revenue"] = (Summary_df["Total Revenue"]).apply(lambda x: '${:,.2f}'.format(x))
Summary_df
```

```python
# Gender Demographics
#find out the unique buyers
Unique_name_total=purchase_data["SN"].value_counts()
Unique_name_total_count=len(Unique_name_total)
#in the unique list find female
female=purchase_data.loc[purchase_data["Gender"] == "Female", :]
Unique_Female=female["SN"].value_counts()
Unique_Female_count=len(Unique_Female)
#in the unique list find male
Male=purchase_data.loc[purchase_data["Gender"] == "Male", :]
Unique_Male=Male["SN"].value_counts()
Unique_Male_count=len(Unique_Male)
#in the unique list find other
Other=Unique_name_total_count-Unique_Female_count-Unique_Male_count
#percentage
female_per=Unique_Female_count/Unique_name_total_count*100
Male_per=Unique_Male_count/Unique_name_total_count*100
other_per=Other/Unique_name_total_count*100
#create df
Gender_df=pd.DataFrame({"":["Male","Female","Other/Non-Disclosed"],
                       "Total Count":[Unique_Male_count,Unique_Female_count,Other],
                       "Percentage of Players":[Male_per,female_per,other_per]})
#formate df
Gender_df["Percentage of Players"]=(Gender_df["Percentage of Players"]).apply(lambda x: '{:,.2f}%'.format(x))
Gender_df
```

```python
#group by name
Top_spender=purchase_data.groupby("SN")
Top_spender_count=Top_spender["Purchase ID"].count()
#Average Purchase Price
Top_spender_avg=Top_spender["Price"].mean()
#Total purchase
Total_top_spender=Top_spender["Price"].sum()
#create df
Top_spender_df=pd.DataFrame({"Purchase Count":Top_spender_count,
                       "Average Purchase Price":Top_spender_avg,
                        "Total Purchase Value":Total_top_spender})
#sort
sort_df=Top_spender_df.sort_values(["Total Purchase Value"], ascending=False)
#formate df
sort_df["Average Purchase Price"]=(sort_df["Average Purchase Price"]).apply(lambda x: '${:,.2f}'.format(x))
sort_df["Total Purchase Value"]=(sort_df["Total Purchase Value"]).apply(lambda x: '${:,.2f}'.format(x))
sort_df.head(6)
```

```python
#group by item id
Top_item=purchase_data.groupby(["Item ID","Item Name"])
Top_item_count=Top_item["SN"].count()
#Item price
Item_price=Top_item["Price"].mean()
#Total purchase
Total_top_item=Top_item["Price"].sum()
#create df
Top_item_df=pd.DataFrame({"Purchase Count":Top_item_count,
                       "Item Price":Item_price,
                        "Total Purchase Value":Total_top_item})
#sort
sort_item_df=Top_item_df.sort_values("Purchase Count", ascending=False)
#formate df
sort_item_df["Total Purchase Value"]=(sort_item_df["Total Purchase Value"]).apply(lambda x: '${:,.2f}'.format(x))
sort_item_df["Item Price"]=(sort_item_df["Item Price"]).apply(lambda x: '${:,.2f}'.format(x))
sort_item_df.head(5)
```
